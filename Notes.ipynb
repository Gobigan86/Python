##### Operators & Operands

```python
### 1.Arithmetic Operators

# Define two variables for demonstration
x = 20
y = 3

print("--- Python Arithmetic Operators Demonstration ---")
print(f"Initial Values: x = {x}, y = {y}\n")

# 1. Addition (+)
addition = x + y
print(f"1. Addition (x + y): {x} + {y} = {addition}") # Output: 23

# 2. Subtraction (-)
subtraction = x - y
print(f"2. Subtraction (x - y): {x} - {y} = {subtraction}") # Output: 17

# 3. Multiplication (*)
multiplication = x * y
print(f"3. Multiplication (x * y): {x} * {y} = {multiplication}") # Output: 60

# 4. Division (/)
# Always returns a float (decimal number), even if the result is an integer.
division = x / y
print(f"4. Division (x / y): {x} / {y} = {division}") # Output: 6.666...

# 5. Modulus (%)
# Returns the remainder after division.
modulus = x % y
print(f"5. Modulus (x % y): {x} % {y} = {modulus}") # Output: 2 (20 divided by 3 is 6 with a remainder of 2)

# 6. Exponentiation (**)
# Raises the first number to the power of the second number (x to the power of y).
exponentiation = x ** y
print(f"6. Exponentiation (x ** y): {x} ** {y} = {exponentiation}") # Output: 8000 (20 * 20 * 20)

# 7. Floor Division (//)
# Divides the numbers and returns the integer part of the quotient (rounds down to the nearest whole number).
floor_division = x // y
print(f"7. Floor Division (x // y): {x} // {y} = {floor_division}") # Output: 6 (The decimal .666 is truncated)

# Example with negative numbers for Floor Division:
a = -20
b = 3
floor_neg = a // b
print(f"   Example: {a} // {b} = {floor_neg}") # Output: -7 (Rounds down to -7, not -6)
```

```python
### 2.Comparison Operators

# Define two variables for comparison
a = 15
b = 10
c = 15

print("--- Python Comparison Operators Demonstration ---")
print(f"Initial Values: a = {a}, b = {b}, c = {c}\n")

# Comparison operators always return a Boolean value (True or False).

# 1. Equal To (==)
# Checks if the two operands are equal.
equal_to = (a == c)
print(f"1. Equal To (a == c): {a} == {c} is {equal_to}") # Output: True
not_equal_to_b = (a == b)
print(f"   Equal To (a == b): {a} == {b} is {not_equal_to_b}") # Output: False

# 2. Not Equal To (!=)
# Checks if the two operands are not equal.
not_equal_to = (a != b)
print(f"2. Not Equal To (a != b): {a} != {b} is {not_equal_to}") # Output: True
equal_to_c = (a != c)
print(f"   Not Equal To (a != c): {a} != {c} is {equal_to_c}") # Output: False

# 3. Greater Than (>)
# Checks if the left operand is greater than the right operand.
greater_than = (a > b)
print(f"3. Greater Than (a > b): {a} > {b} is {greater_than}") # Output: True

# 4. Less Than (<)
# Checks if the left operand is less than the right operand.
less_than = (b < a)
print(f"4. Less Than (b < a): {b} < {a} is {less_than}") # Output: True

# 5. Greater Than or Equal To (>=)
# Checks if the left operand is greater than or equal to the right operand.
greater_or_equal_1 = (a >= b)
greater_or_equal_2 = (a >= c)
print(f"5. Greater Than or Equal To:")
print(f"   (a >= b): {a} >= {b} is {greater_or_equal_1}") # Output: True
print(f"   (a >= c): {a} >= {c} is {greater_or_equal_2}") # Output: True

# 6. Less Than or Equal To (<=)
# Checks if the left operand is less than or equal to the right operand.
less_or_equal_1 = (b <= a)
less_or_equal_2 = (a <= c)
print(f"6. Less Than or Equal To:")
print(f"   (b <= a): {b} <= {a} is {less_or_equal_1}") # Output: True
print(f"   (a <= c): {a} <= {c} is {less_or_equal_2}") # Output: True
```

```python
### 3.Logical Operators

# Define a few Boolean variables to use in the examples
is_raining = True
is_cold = False
age = 25
is_student = True

print("--- Python Logical Operators Demonstration ---")
print(f"Initial State: is_raining={is_raining}, is_cold={is_cold}, age={age}, is_student={is_student}\n")

# 1. Logical AND (and)
# The result is TRUE only if BOTH conditions are TRUE.
condition_and_1 = is_raining and is_cold
print(f"1. AND (is_raining and is_cold): {is_raining} and {is_cold} = {condition_and_1}")
# Expected Output: False (Since is_cold is False)

# Example combining comparison and logic
is_adult_and_student = (age >= 18) and is_student
print(f"   AND ((age >= 18) and is_student): {is_adult_and_student}")
# Expected Output: True (Since 25 >= 18 is True, and is_student is True)


# 2. Logical OR (or)
# The result is TRUE if AT LEAST ONE condition is TRUE.
condition_or_1 = is_raining or is_cold
print(f"2. OR (is_raining or is_cold): {is_raining} or {is_cold} = {condition_or_1}")
# Expected Output: True (Since is_raining is True)

condition_or_2 = (age < 18) or is_cold
print(f"   OR ((age < 18) or is_cold): {(age < 18)} or {is_cold} = {condition_or_2}")
# Expected Output: False (Since both (age < 18) and is_cold are False)


# 3. Logical NOT (not)
# The result inverts the Boolean value of the condition (True becomes False, and False becomes True).
condition_not_1 = not is_raining
print(f"3. NOT (not is_raining): not {is_raining} = {condition_not_1}")
# Expected Output: False

condition_not_2 = not is_cold
print(f"   NOT (not is_cold): not {is_cold} = {condition_not_2}")
# Expected Output: True

# Complex Example: It's raining OR it's cold, BUT you must NOT be a student.
complex_condition = (is_raining or is_cold) and (not is_student)
print(f"\n4. Complex Condition:")
print(f"   (is_raining or is_cold) and (not is_student) = {complex_condition}")
# Breakdown: (True or False) -> True. AND (not True) -> False.
# Expected Output: False
```

```python
### 4.Assignment Operators

# Define a starting variable
x = 10

print("--- Python Assignment Operators Demonstration ---")
print(f"Starting Value: x = {x}\n")

# 1. Simple Assignment (=)
# Assigns the value on the right to the variable on the left.
y = 5
print(f"1. Simple Assignment (y = 5): y is now {y}")
# Note: This is used for all subsequent operators.

# Reset x for the next test
x = 10

# 2. Add AND Assignment (+=)
# x = x + 3
x += 3
print(f"2. Add AND Assignment (x += 3): x is now {x}") # Output: 13 (10 + 3)

# 3. Subtract AND Assignment (-=)
# x = x - 5
x -= 5
print(f"3. Subtract AND Assignment (x -= 5): x is now {x}") # Output: 8 (13 - 5)

# 4. Multiply AND Assignment (*=)
# x = x * 2
x *= 2
print(f"4. Multiply AND Assignment (x *= 2): x is now {x}") # Output: 16 (8 * 2)

# 5. Divide AND Assignment (/=)
# x = x / 4 (The result is a float)
x /= 4
print(f"5. Divide AND Assignment (x /= 4): x is now {x}") # Output: 4.0 (16 / 4)

# Reset x to an integer for the next tests
x = 10

# 6. Modulus AND Assignment (%=)
# x = x % 3 (Remainder of 10 / 3)
x %= 3
print(f"6. Modulus AND Assignment (x %= 3): x is now {x}") # Output: 1 (10 % 3)

# Reset x
x = 2

# 7. Exponentiation AND Assignment (**=)
# x = x ** 3 (2 to the power of 3)
x **= 3
print(f"7. Exponentiation AND Assignment (x **= 3): x is now {x}") # Output: 8 (2 * 2 * 2)

# Reset x
x = 15

# 8. Floor Division AND Assignment (//=)
# x = x // 4 (Integer quotient of 15 / 4)
x //= 4
print(f"8. Floor Division AND Assignment (x //= 4): x is now {x}") # Output: 3 (15 // 4)
```

```python
### 5.Identity Operators

# --- Python Identity Operators Demonstration ---

# 1. Basic Integer Comparison
# Python often optimizes small integers by creating only one copy in memory.
a = 10
b = 10
c = 20

print("1. Basic Integer Comparison:")
print(f"a = {a}, b = {b}, c = {c}")

# 'is' checks if a and b refer to the exact same object in memory.
print(f"a is b: {a is b}")        # Output: True (Python optimization)
print(f"a is not c: {a is not c}") # Output: True

# Note: a == b checks for value equality, which is also True.
print(f"a == b: {a == b}\n")

# 2. List Comparison (Lists are Mutable)
# Lists are mutable, meaning they are rarely optimized. New lists are almost always new objects.
list1 = [1, 2, 3]
list2 = [1, 2, 3] # New list object, even with the same content
list3 = list1     # list3 refers to the SAME object as list1

print("2. List Comparison (Mutable Objects):")
print(f"list1: {list1}, list2: {list2}, list3: {list3}")

# 'is' compares memory identity
print(f"list1 is list2: {list1 is list2}")   # Output: False (Different objects in memory)
print(f"list1 == list2: {list1 == list2}")   # Output: True (Same content/value)
print(f"list1 is list3: {list1 is list3}")   # Output: True (They point to the same object)

# 'is not' checks for different memory identity
print(f"list1 is not list2: {list1 is not list2}\n") # Output: True

# 3. String Comparison (Strings are Immutable)
# Strings are immutable. For short, identical strings, Python often uses memory optimization.
str1 = "hello"
str2 = "hello"
str3 = "world"

print("3. String Comparison (Immutable Objects):")
print(f"str1: '{str1}', str2: '{str2}', str3: '{str3}'")

print(f"str1 is str2: {str1 is str2}")     # Output: True (Likely due to memory optimization/interning)
print(f"str1 is not str3: {str1 is not str3}\n") # Output: True


# --- Summary of Identity Operators ---

# 'is'        -> True if both variables are the SAME object (same memory ID).
# 'is not'    -> True if both variables are NOT the same object (different memory IDs).
```

```python
### 6.Bitwise Operators

# --- Python Bitwise Operators Demonstration ---

# Define two integers for demonstration.
# Their binary representations are:
# A = 10 (Binary: 0000 1010)
# B = 4  (Binary: 0000 0100)
A = 10
B = 4
C = -11 # Used for the NOT operator example

print(f"A (Decimal): {A} | Binary: {bin(A)}")
print(f"B (Decimal): {B} | Binary: {bin(B)}")
print("-" * 40)

# Helper function to print results clearly
def print_bitwise_result(op_name, op_symbol, result):
    print(f"{op_name} ({op_symbol}):")
    print(f"  Result (Decimal): {result}")
    print(f"  Result (Binary): {bin(result)}\n")


# 1. Bitwise AND (&)
# Sets each bit to 1 if BOTH bits are 1.
#   A: 1010
# & B: 0100
# -----
#   R: 0000
result_and = A & B
print_bitwise_result("1. AND", "&", result_and)


# 2. Bitwise OR (|)
# Sets each bit to 1 if AT LEAST ONE of the two bits is 1.
#   A: 1010
# | B: 0100
# -----
#   R: 1110 (14 in decimal)
result_or = A | B
print_bitwise_result("2. OR", "|", result_or)


# 3. Bitwise XOR (^)
# Sets each bit to 1 if ONLY ONE of the two bits is 1 (exclusive OR).
#   A: 1010
# ^ B: 0100
# -----
#   R: 1110 (14 in decimal)
result_xor = A ^ B
print_bitwise_result("3. XOR", "^", result_xor)



# 4. Bitwise NOT (~)
# Inverts all bits (flips 0s to 1s and 1s to 0s).
# Note: For positive numbers, the result is typically -(N + 1).
# Python uses two's complement for negative numbers.
# For C = 10: result is -(10 + 1) = -11
result_not = ~A
print_bitwise_result("4. NOT", "~", result_not)
print(f"   For C = -11: ~C = {~C} (i.e., -(-11 + 1) = 10)\n")


# 5. Bitwise Left Shift (<<)
# Shifts the bits to the left, adding 0s on the right.
# Shifting by N positions is equivalent to multiplying by 2**N.
# A = 10 (1010) shifted left by 2 (101000) = 40
result_left_shift = A << 2
print_bitwise_result("5. Left Shift", "<< 2", result_left_shift)


# 6. Bitwise Right Shift (>>)
# Shifts the bits to the right, discarding the bits on the right.
# Shifting by N positions is equivalent to floor dividing by 2**N.
# A = 10 (1010) shifted right by 2 (0010) = 2
result_right_shift = A >> 2
print_bitwise_result("6. Right Shift", ">> 2", result_right_shift)
```

```python
### 7.Membership operators

# --- Python Membership Operators Demonstration ---

# 1. Using a List
my_list = ['apple', 'banana', 'cherry', 'date']
search_item = 'banana'
missing_item = 'grape'

print("--- 1. Testing Membership in a List ---")
print(f"List: {my_list}")

# 'in' operator
# Check if 'banana' is in the list
print(f"Is '{search_item}' in the list? {'banana' in my_list}") # Output: True

# Check if 'grape' is in the list
print(f"Is '{missing_item}' in the list? {'grape' in my_list}") # Output: False

# 'not in' operator
# Check if 'grape' is NOT in the list
print(f"Is '{missing_item}' NOT in the list? {'grape' not in my_list}") # Output: True
print("-" * 40)


# 2. Using a String
my_string = "Python is powerful"
search_char = 'p' # Case sensitive!
search_word = 'is'
missing_word = 'java'

print("--- 2. Testing Membership in a String ---")
print(f"String: '{my_string}'")

# 'in' checks for substrings
print(f"Is 'p' in the string? {'p' in my_string}")       # Output: False (Case sensitive)
print(f"Is 'P' in the string? {'P' in my_string}")       # Output: True
print(f"Is '{search_word}' in the string? {'is' in my_string}") # Output: True

# 'not in' operator
print(f"Is '{missing_word}' NOT in the string? {'java' not in my_string}") # Output: True
print("-" * 40)


# 3. Using a Dictionary
# Membership checks for DICTIONARIES ONLY check the KEYS, not the values.
my_dict = {
    'name': 'Alice',
    'age': 30,
    'city': 'London'
}

print("--- 3. Testing Membership in a Dictionary (Keys Only) ---")
print(f"Dictionary: {my_dict}")

# Checking for a key
print(f"Is 'age' a key in the dictionary? {'age' in my_dict}") # Output: True

# Checking for a value (This will be FALSE!)
print(f"Is 'Alice' a key in the dictionary? {'Alice' in my_dict}") # Output: False

# 'not in' operator
# Check if 'country' is NOT a key
print(f"Is 'country' NOT a key? {'country' not in my_dict}") # Output: True
```

##### Conditional Statement

```python
# --- Python Conditional Statements Demo ---

# Scenario: Determine a user's status based on their age.

age = 17
print(f"Current Age: {age}")
print("-" * 30)

# The IF statement: Checks the first condition.
if age >= 18:
    # This block executes ONLY if the condition is True.
    status = "Adult (Eligible to vote)"
    print("Condition 1 (age >= 18) is TRUE.")

# The ELIF statement (Else If): Checks additional conditions only if the preceding 'if'/'elif' conditions were False.
elif age >= 13:
    # This block executes if the first condition was False AND this condition is True.
    status = "Teenager (Requires parental consent)"
    print("Condition 2 (age >= 13) is TRUE.")

# The ELSE statement: The catch-all.
else:
    # This block executes ONLY if ALL preceding 'if' and 'elif' conditions were False.
    status = "Child (Minor)"
    print("All preceding conditions are FALSE.")


print("-" * 30)
# Output the final determined status
print(f"Calculated Status: {status}")

# --- Example 2: Using Boolean Values ---
is_logged_in = True
item_price = 150

if is_logged_in and item_price > 100:
    print("\nUser is logged in and the item is expensive.")
    discount = 0.10
elif is_logged_in:
    print("\nUser is logged in but the item is inexpensive.")
    discount = 0.05
else:
    print("\nUser is NOT logged in. No discount.")
    discount = 0.0
    
print(f"Applied Discount: {discount * 100:.0f}%")

## shorthand if else 
marks = 87
print("You will go to the trip") if marks >= 89 else print("You will not go to the trip")
```

##### built-in string functions in Python used for validation


```python
s_alnum = "Python32"       # Alphanumeric (letters and numbers)
s_alpha = "BeautifulSoup"  # Alphabetic (letters only)
s_decimal = "12345"        # Decimal characters (0-9)
s_digit = "\u00B2"         # Digit characters (0-9, subscripts, superscripts, etc.)
s_numeric = "123.45"       # Numeric characters (0-9, fractions, decimals, etc.)
s_lower = "example"        # Lowercase only
s_upper = "PYTHON"         # Uppercase only
s_space = "   "            # Whitespace only
s_title = "The Quick Brown Fox" # Titlecase (first letter of each word is capitalized)
s_mixed = "Abc 123!"       # Mixed content (contains non-alphanumeric, spaces)


print("--- Validation Results ---")
print(f"{'Method':<12} | {'Description':<28} | {'Test String':<18} | Result")
print("-" * 65)



print(s_mixed.isalnum())

# 2. isalpha(): Checks if all characters are alphabetic (A-z only)
print(f"{'isalpha()':<12} | {'Alphabetic (letters) only':<28} | {s_alpha:<18} | {s_alpha.isalpha()}") # True
print(f"{'':<12} | {'':<28} | {s_alnum:<18} | {s_alnum.isalpha()}") # False (because of '32')


# 3. isdecimal(): Checks if all characters are decimal digits (0-9)
# Note: This often only returns true for standard ASCII digits.
print(f"{'isdecimal()':<12} | {'Decimal digits (0-9) only':<28} | {s_decimal:<18} | {s_decimal.isdecimal()}") # True
print(f"{'':<12} | {'':<28} | {s_digit:<18} | {s_digit.isdecimal()}") # False (Superscript 2 is not a decimal)


# 4. isdigit(): Checks if all characters are digits (0-9, and some unicode digits)
print(f"{'isdigit()':<12} | {'Digits (0-9) + Unicode digits':<28} | {s_decimal:<18} | {s_decimal.isdigit()}") # True
print(f"{'':<12} | {'':<28} | {s_digit:<18} | {s_digit.isdigit()}") # True (Superscript 2 \u00B2 is a digit)


# 5. isnumeric(): Checks if all characters are numeric (the broadest category)
# Includes digits, fractions, superscripts, and other forms of numbers.
print(f"{'isnumeric()':<12} | {'Numeric (broadest category)':<28} | {'1/2':<18} | {'1/2'.isnumeric()}") # False (Slash is not numeric)
print(f"{'':<12} | {'':<28} | {'u00BD':<18} | {'u00BD'.isnumeric()}") # True (Unicode Fraction 1/2 is numeric)
print(f"{'':<12} | {'':<28} | {s_numeric:<18} | {s_numeric.isnumeric()}") # False (because of the decimal point '.')


# 6. islower(): Checks if all cased characters are lowercase
print(f"{'islower()':<12} | {'All cased characters are lowercase':<28} | {s_lower:<18} | {s_lower.islower()}") # True
print(f"{'':<12} | {'':<28} | {'Test string':<18} | {'Test string'.islower()}") # False (because of 'T')


# 7. isupper(): Checks if all cased characters are uppercase
print(f"{'isupper()':<12} | {'All cased characters are uppercase':<28} | {s_upper:<18} | {s_upper.isupper()}") # True
# isupper() ignores non-cased characters like numbers and symbols, only checking letters
print(f"{'':<12} | {'':<28} | {'DATA101':<18} | {'DATA101'.isupper()}") # True


# 8. isspace(): Checks if all characters in the string are whitespace
print(f"{'isspace()':<12} | {'All characters are whitespace':<28} | {s_space:<18} | {s_space.isspace()}") # True
print(f"{'':<12} | {'':<28} | {' a ':<18} | {' a '.isspace()}") # False (because of 'a')


# 9. istitle(): Checks if the string is correctly title-cased
# Every word must start with an uppercase letter, and all other letters must be lowercase.
print(f"{'istitle()':<12} | {'Follows title case rules':<28} | {s_title:<18} | {s_title.istitle()}") # True
print(f"{'':<12} | {'':<28} | {'a simple title':<18} | {'a simple title'.istitle()}") # False (because of 'a')
```

### sets - unordered collection of data

```python
a = {"alice", "bob", "charlie","ram","ravi","bob"}
print(a)
print(type(a))


### Functions

#add
a.add("david")
print(a)

#pop
a.pop()
print(a)

#remove
a.remove("ram")         ## if the element is not present , it will raise an error
print("Remove",a)

#discard
a.discard("ravi1")              ## if the element is not present , it will not raise an error
print(a)    

#clear
a.clear()
print(a)

#isdisjoint
set1 = {1,2,3}
set2 = {4,5,6}
set3 = {3,4,5}
print(set1.isdisjoint(set2))   
print(set1.isdisjoint(set3))   

#issubset
setA = {1,2}
setB = {1,2,3,4,5}
print(setA.issubset(setB))  

#superset
setC = {1,2,3,4,5}
setD = {2,3}
print(setC.issuperset(setD))

#update
setC.update({6,7})
print(setC)

#union
setE = setA.union(setB)
print(setE)

#clear
setE.clear()
print(setE)

#intersection
setF = {1,2,3,4}
setG = {3,4,5,6}
setH = setF.intersection(setG)
print(setH)

#difference
setI = setF.difference(setG)
print(setI)

#symmetric_difference
setJ = setF.symmetric_difference(setG)
print(setJ)

#symmetric_difference_update
setF.symmetric_difference_update(setG)
print(setF)
```

```python
def factorial(n):
    if n==1:
        return 1
    else:
        return (n*factorial(n-1))

print(f"factorial of {5} is {factorial(5)}")
```
